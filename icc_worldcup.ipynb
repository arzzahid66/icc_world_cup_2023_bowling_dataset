```python

avg_runs = icc.groupby('team')['runs'].mean().reset_index()
print(avg_runs)
```

```python
team_colors = ['black', 'yellow', '#006600', 'red', '#ff9933', 'blue', 'silver', '#00FF00', 'gold', 'maroon']
```

```python

avg_runs = icc.groupby('team')['runs'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x="runs", y="team", data=avg_runs)
plt.title('Average Runs by Team')
plt.xlabel('Average Runs')
plt.ylabel('Team')
plt.show()
```

```python
# teams vs wickets
plot = sns.barplot(y="team",x="wickets", data=icc)
```

```python


wickets_by_team = icc.groupby('team')['wickets'].sum()

plt.pie(wickets_by_team,
        labels=wickets_by_team.index,
        colors=team_colors,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Wickets Distribution by Team')
plt.show()

```

```python
wickets_by_team = icc.groupby('team')['6s'].sum()
plt.pie(wickets_by_team, labels=wickets_by_team.index, autopct='%1.1f%%', startangle=90 , colors=team_colors)
plt.title('sixs Distribution by Team')
plt.show()
```

```python
wickets_by_team = icc.groupby('team')['4s'].sum()
plt.pie(wickets_by_team, labels=wickets_by_team.index, autopct='%1.1f%%', startangle=90,colors=team_colors)
plt.title('Fours Distribution by Team')
plt.show()
```

```python
# Assuming your DataFrame is named 'icc'
top_players = icc.groupby('player')['wickets'].sum().sort_values(ascending=False).head(5)
sns.barplot(x=top_players.index, y=top_players.values)
plt.xticks(rotation=45, ha='right')
plt.title('Top 5 Players with Most Wickets')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Given data
player_data = {'Player': ['Mohammed Shami', 'Adam Zampa', 'Dilshan Madushanka', 'Jasprit Bumrah', 'Gerald Coetzee'],
               'Wickets': [24, 23, 21, 20, 20]}

# Create a DataFrame
df = pd.DataFrame(player_data)

# Sort the DataFrame by Wickets in descending order
df = df.sort_values(by='Wickets', ascending=False)

# Define custom colors for each player
custom_colors = sns.color_palette("husl", n_colors=len(df))

# Use Seaborn to create a horizontal bar plot with custom colors
plt.figure(figsize=(8,4))
sns.barplot(x='Wickets', y='Player', data=df, palette=custom_colors)

# Adding labels and title
plt.xlabel('Total Wickets')
plt.ylabel('Players')
plt.title('Top Players by Total Wickets')

# Display the plot
plt.show()
```
